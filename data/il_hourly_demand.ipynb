```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./demand_profile.csv")
df
```

```python
seasons = np.unique(df.season)

fig, ax = plt.subplots()
fig.figsize=(12,9)
for season in seasons:
    ax.plot(range(24), df[df.season == season].demand_frac, label=season.capitalize())
plt.xlabel('Hour')
ax.legend()
ax.set_ylabel('Fraction of Total Annual Demand')
plt.show()
```
